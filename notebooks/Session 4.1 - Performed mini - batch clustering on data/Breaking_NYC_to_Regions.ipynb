```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_reader = pd.read_csv(data_path, chunksize=100000)
```

```python
# train the standard scaler

scaler = StandardScaler()

for chunk in df_reader:
    # fit the scaler
    scaler.partial_fit(chunk)
```

**Why Use partial_fit()?**
- Uses partial_fit(chunk) instead of fit(), which allows the scaler to process large datasets incrementally.
- Unlike fit(), partial_fit() does not require loading the entire dataset into memory.
- It updates mean and variance using streaming data.

```python
def haversine_distance(lat1: float, lon1:float, lat2:float, lon2:float):
    
    """Calculate haversine distances between two points given
    their latitude and longitude coordinates.
    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = np.sin(
        dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    
    c = 2*np.arcsin(np.sqrt(a))
    
    earth_radius =3958.8
    distance = earth_radius * c
    return distance
```

```python
#Function Definition and Loop Over k_values
def select_optimum_k(k_values):
    for k in k_values:
        
        # Read Data in Chunks (Handles Large Datasets)
        df_reader = pd.read_csv(data_path, chunksize=100000)
        
        # Initialize MiniBatch K-Means Model
        mini_batch = MiniBatchKMeans(n_clusters=k, n_init=10, random_state=42)
        
        #Train the Model on Scaled Data
        for chunk in df_reader:
            # scale the chunk: Normalizes (scales) the chunk using scaler.transform().
            scaled_chunk = scaler.transform(chunk)
            # train the model: Trains K-Means incrementally using partial_fit(), updating cluster centers.
            mini_batch.partial_fit(scaled_chunk)
            
        # centroid values: Convert Cluster Centers Back to Original Scale
        centroids = scaler.inverse_transform(mini_batch.cluster_centers_)
        
        
        # Compute Pairwise Cluster Distances
        all_distances = []
        for i in centroids:
            distances = []
            for j in centroids:
                distance = haversine_distance(i[0], i[1], j[0], j[1])
                distances.append(distance)
            all_distances.append(distances)
        all_distances = np.array(all_distances)
        
        
        # Sorts distances row-wise, so that the closest distances appear first.
        sorted_distances = np.sort(all_distances, axis=-1)
        
        # Extract Distances to the 8 Nearest Clusters
        # nearby regions
        region_count = 8
        # selected distances
        selected_distances = sorted_distances[:, 1:region_count+1] # Skips [:, 0] because it is always 0 (distance to itself).

        # Computes the average distance to the 8 nearest clusters.
        avg_distances = selected_distances.mean(axis=1)
        
        # count averages that fit criteria
        counter = ((avg_distances >= 1.0) & (avg_distances <= 1.5)).sum()
        percentage = ((avg_distances >= 1.0) & (avg_distances <= 1.5)).mean()
        print(f"For {k} clusters, the number of nearby regions that fits criteria are {counter} and is {percentage * 100:.1f}% regions of total")
```

```python
possible_k_values = np.arange(10, 100, 10)
possible_k_values
```

```python
# select the optimum value of K

select_optimum_k(k_values=possible_k_values)
```
